```python
import requests
import pandas as pd

# List of API URLs

#Carbon footprint of bank loans - climate change
#Environmental tax -mitigation
#Environmental protection expenditure -mitigation
#Fossil fuel subsidiaries - mitigation
api_urls = [
    "https://services9.arcgis.com/weJ1QsnbMYJlCHdG/arcgis/rest/services/Indicator_14_Expenditure_on_Environmental_Protection/FeatureServer/0/query?where=1%3D1&outFields=*&outSR=4326&f=json",
    "https://services9.arcgis.com/weJ1QsnbMYJlCHdG/arcgis/rest/services/Indicator_13_Environmental_Taxes_new/FeatureServer/0/query?where=1%3D1&outFields=*&outSR=4326&f=json",
    "https://services9.arcgis.com/weJ1QsnbMYJlCHdG/arcgis/rest/services/Indicator_9_Carbon_footprint_adjusted_loans_to_total_loans_for_deposit_takers/FeatureServer/0/query?where=1%3D1&outFields=*&outSR=4326&f=json",
    "https://services9.arcgis.com/weJ1QsnbMYJlCHdG/arcgis/rest/services/Indicator 15 Fossil Fuel Subsidies/FeatureServer/0/query?where=1%3D1&outFields=*&outSR=4326&f=json"
]

def fetch_and_union(api_urls):
    dfs = []

    for url in api_urls:
        # Fetch data from the API
        response = requests.get(url)
        if response.status_code == 200:
            json_data = response.json()

            # Extract 'attributes' from 'features'
            features = json_data.get("features", [])
            attributes = [feature["attributes"] for feature in features]

            # Convert to DataFrame
            df = pd.DataFrame(attributes)
            # Rename columns like 'F2021' to '2021' only if it starts with 'F' and is followed by digits
            df.rename(
                columns=lambda col: col[1:] if col.startswith("F") and col[1:].isdigit() else col,
                inplace=True
            )
            dfs.append(df)
        else:
            print(f"Failed to fetch data from {url}, Status Code: {response.status_code}")

    # Handle empty dfs list
    if not dfs:
        return pd.DataFrame()

    # Union all DataFrames and handle extra columns
    all_columns = set(col for df in dfs for col in df.columns)
    aligned_dfs = [df.reindex(columns=all_columns, fill_value=None) for df in dfs]

    # Concatenate DataFrames
    final_df = pd.concat(aligned_dfs, ignore_index=True)
    return final_df

# Execute the function
final_dataframe = fetch_and_union(api_urls)

# Display the final DataFrame
print(final_dataframe.columns )
```
